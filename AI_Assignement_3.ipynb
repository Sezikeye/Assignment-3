```python
def check_availability(book_name):
    books = ["Python for Beginners","Python Basics", "Advanced Python", "Data Science with Python","Machine Learning"]

    if book_name in books:
        return True
    else:
        return False
```

```python
def save_books_to_file(filename, books):
    with open(filename, 'w') as f:
        for book in books:
            f.write(f"{book}\n")
```

```python
def read_books_from_file(filename):
    books = []

    with open(filename, 'r') as f:
        for line in f:
            books.append(line.strip())

    return books
```

```python
books = ["Python for Beginners","Python Basics", "Advanced Python", "Data Science with Python","Machine Learning"]

save_books_to_file("books.txt", books)
!cat books.txt

```

```python
def save_books_pickle(filename, books):
    with open(filename, 'wb') as f:
        pickle.dump(books, f)
```

```python
def load_books_pickle(filename):
    with open(filename, 'rb') as f:
        books = pickle.load(f)
    return books
```

```python
books = ["Python for Beginners","Python Basics", "Advanced Python", "Data Science with Python","Machine Learning"]
save_books_pickle("books.pkl", books)
books_loaded = load_books_pickle("books.pkl")
print(books_loaded)
```

```python
def create_books_table():
    conn = sqlite3.connect('library.db')
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS books
                      (id INTEGER PRIMARY KEY, title TEXT, quantity INTEGER)''')
    conn.commit()
    conn.close()
```

```python
import sqlite3

def add_book_to_db(title, quantity):
    conn = sqlite3.connect('library.db')
    cursor = conn.cursor()

    cursor.execute(
        "INSERT INTO books (title, quantity) VALUES (?, ?)",
        (title, quantity)
    )

    conn.commit()
    conn.close()
```

```python
create_books_table()
add_book_to_db("Python Basics", 5)
add_book_to_db("Advanced Python", 3)
add_book_to_db("Data Science with Python", 2)
```

```python
def get_python_repos():
    url = "https://api.github.com/search/repositories"
    params = {"q": "language:python", "sort": "stars", "order": "desc"}
    response = requests.get(url, params=params)
    data = response.json()
    return [repo['name'] for repo in data['items'][:5]]
```

```python
print("Top 5 Python repositories on GitHub:", get_python_repos())
```
